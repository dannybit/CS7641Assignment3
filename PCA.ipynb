```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import pandas as pd
from collections import Counter
import numpy as np
from sklearn.metrics import accuracy_score
from sklearn.metrics import adjusted_mutual_info_score as ami, silhouette_score as sil_score, silhouette_samples as sil_samples
from sklearn.mixture import GaussianMixture
from collections import defaultdict
import matplotlib.pyplot as plt
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.decomposition import PCA
from helpers import cluster_acc
from sklearn.neural_network import MLPClassifier
from time import clock
```

**Load Spam**

```python
spam_df = pd.read_csv('spambase.csv', header=0)
spam_x = spam_df.drop('class', axis=1).copy().values
spam_y = spam_df['class'].copy().values
spam_x = StandardScaler().fit_transform(spam_x)
spam_x.shape
```

**Load Digits**

```python
from sklearn.datasets import load_digits
digits = load_digits()
digits_x = StandardScaler().fit_transform(digits.data)
digits_y = digits.target
digits_x.shape
```

```python
spamPca = PCA(random_state=5)
spamPca.fit(spam_x)
```

```python
pd.Series(np.cumsum(spamPca.explained_variance_ratio_)).plot(title="Spam PCA explained variance ratio")
```

```python
digitsPca = PCA(random_state=5)
digitsPca.fit(digits_x)
```

```python
pd.Series(np.cumsum(digitsPca.explained_variance_ratio_)).plot(title="Digits PCA explained variance ratio")
```

```python
from numpy import linalg as LA
spam_rec_error = []
digits_rec_error = []
for i in range(1, spam_x.shape[1] + 1):
    pca = PCA(random_state=5, n_components=i)
    transformed = pca.fit_transform(spam_x)
    inversed_transformed = pca.inverse_transform(transformed)
    spam_rec_error.append(LA.norm((spam_x-inversed_transformed),None))
```

```python
digits_rec_error = []
for i in range(1, digits_x.shape[1] + 1):
    pca = PCA(random_state=5, n_components=i)
    transformed = pca.fit_transform(digits_x)
    inversed_transformed = pca.inverse_transform(transformed)
    digits_rec_error.append(LA.norm((digits_x-inversed_transformed),None))
```

```python
pd.Series(spam_rec_error).plot(title='Spam PCA reconstruction error')
```

```python
pd.Series(digits_rec_error).plot(title='Digits PCA reconstruction error')
```

```python
spam_pca = PCA(random_state=5, n_components=40)
spam_x_red = spam_pca.fit_transform(spam_x)
digits_pca = PCA(random_state=5, n_components=20)
digits_x_red = digits_pca.fit_transform(digits_x)
```

```python
km = KMeans(random_state=5)
gmm = GaussianMixture(random_state=5)
clusters =  [2,5,10,15,20,25,30,35,40,50,70,80,90,100,115,130]
accSpam = {}
acc = defaultdict(lambda: defaultdict(dict))
SSE = defaultdict(dict)
ll = defaultdict(dict)
for k in clusters:
    km.set_params(n_clusters=k)
    gmm.set_params(n_components=k)
    km.fit(digits_x_red)
    gmm.fit(digits_x_red)
    acc[k]['digits']['kmeans'] = cluster_acc(digits_y, km.predict(digits_x_red))
    acc[k]['digits']['gmm'] = cluster_acc(digits_y, gmm.predict(digits_x_red))
    SSE[k]['digits'] = km.score(digits_x_red)
    ll[k]['digits'] = gmm.score(digits_x_red)
    km.fit(spam_x_red)
    gmm.fit(spam_x_red)
    acc[k]['spam']['kmeans'] = cluster_acc(spam_y, km.predict(spam_x_red))
    acc[k]['spam']['gmm'] = cluster_acc(spam_y, gmm.predict(spam_x_red))
    SSE[k]['spam'] = km.score(spam_x_red)
    ll[k]['spam'] = gmm.score(spam_x_red)    
```

```python
kmeansAccSpam = defaultdict(dict)
kmeansAccDigits = defaultdict(dict)
for k in acc:
    kmeansAccSpam[k] = acc[k]['digits']['kmeans']
    kmeansAccDigits[k] = acc[k]['spam']['kmeans']

pd.DataFrame.from_dict(kmeansAccSpam, orient='index', columns=['acc']).plot(title='kmeans acc spam pca reduced', legend=False)
pd.DataFrame.from_dict(kmeansAccDigits, orient='index', columns=['acc']).plot(title='kmeans acc digits pca reduced', legend=False)
```

```python
pd.DataFrame.from_dict(SSE, orient='index')['spam'].plot(title='spam SSE pca reduced')
plt.show()
pd.DataFrame.from_dict(SSE, orient='index')['digits'].plot(title='digits SSE pca reduced')
```

```python
pd.DataFrame.from_dict(ll, orient='index')['digits'].plot(title='digits ll pca reduced')
plt.show()
pd.DataFrame.from_dict(ll, orient='index')['spam'].plot(title='spam ll pca reduced')
```

```python
gmmAccSpam = defaultdict(dict)
gmmAccDigits = defaultdict(dict)
for k in acc:
    gmmAccSpam[k] = acc[k]['digits']['gmm']
    gmmAccDigits[k] = acc[k]['spam']['gmm']

pd.DataFrame.from_dict(gmmAccSpam, orient='index', columns=['acc']).plot(title='gmm acc spam', legend=False)
pd.DataFrame.from_dict(gmmAccDigits, orient='index', columns=['acc']).plot(title='gmm acc digits', legend=False)
```

```python
from sklearn.metrics import make_scorer, accuracy_score
dims_times = {}
dims_acc = {}
for d in range(1, spam_x.shape[1] + 1):
    pca = PCA(random_state=5, n_components=d)
    spam_red = pca.fit_transform(spam_x)
    mlp = MLPClassifier(activation='relu',max_iter=2000,early_stopping=True,random_state=5, hidden_layer_sizes=(spam_red.shape[1], 57, 57))
    st = clock()
    mlp.fit(spam_red, spam_y)
    dims_times[d] = clock() - st
    dims_acc[d] = accuracy_score(mlp.predict(spam_red), spam_y)
```

```python
pd.DataFrame.from_dict(dims_acc, orient='index').plot(title='NN pca spam acc', legend=False)
pd.DataFrame.from_dict(dims_times, orient='index').plot(title='NN pca spam times', legend=False)
```
